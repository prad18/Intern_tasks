##  TASK 1: EXTRACT NAMED ENTITIES FROM PDF (SPACY LAYOUT)


```python
import spacy
from spacy_layout import spaCyLayout
```

```python
nlp = spacy.load("en_core_web_trf")
```

```python
layout = spaCyLayout(nlp)

doc = layout("./stories.pdf")
```

```python
doc = nlp(doc)
doc
```

```python
sentence1=list(doc.sents)
sentence1
```

```python
for ent in doc.ents:
    stuct_format= f"TEXT : {ent.text}  NER: {ent.label_}"
    print(stuct_format)
```

```python
with open("stories.txt",'w') as f:
    for ent in doc.ents:
        f.write(f"TEXT : {ent.text}  NER: {ent.label_}\n")
```

```python
from spacy import displacy
displacy.render(doc,style="ent")
```
